```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

### Logarithmic least squares fittings
Via [Online Logarithmic Regression](http://www.xuru.org/rt/LnR.asp) tool.

```python
def Q_s(p):
    return 168.8415387 * np.log(p) - 6.509428059

def Q_d(p):
    return -166.9211364 * np.log(p) + 266.7766827
```

## Demand & supply functions. Equilibrium

```python
# Points, demand & supply functions
p = np.arange(0.01, 5.0, 0.01)

plt.plot([280, 245, 190, 141, 135, 110, 95, 65, 58, 44, 21, 10], 
         [1, 1.25, 1.57, 1.81, 2.09, 2.45, 2.8, 3.19, 3.58, 3.85, 4.5, 5], 'ro')
plt.plot(Q_s(p), p, lw=2, c='blue')

plt.plot([5, 20, 51, 89, 120, 153, 180, 201, 215, 228, 240, 248],
         [1, 1.25, 1.57, 1.81, 2.09, 2.45, 2.8, 3.19, 3.58, 3.85, 4.5, 5], 'bo')
plt.plot(Q_d(p), p, lw=2, c='red')

# Equilibrium
#idx = np.argwhere(np.diff(np.sign(Q_s(p) - Q_d(p))) != 0).reshape(-1) + 0
#plt.plot(Q_s(p)[idx], p[idx], 'go')
plt.annotate('Equilibrium', xy=(130.915, 2.25675), xytext=(125, 3),
            arrowprops=dict(facecolor='black', shrink=0.05))


# Legend
supply_patch = mpatches.Patch(color='b', label='Supply')
demand_patch = mpatches.Patch(color='r', label='Demand')
plt.legend(handles=[supply_patch, demand_patch])


# Plot scaling
plt.title('MMEEP lab1')
plt.xlabel('Quantity, N')
plt.ylabel('Price, $')
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 20
fig_size[1] = 10
plt.rcParams["figure.figsize"] = fig_size
plt.grid(True)
plt.xlim([0, 300])
plt.ylim([0, 6])
plt.show()
```

## Supply function with subsidy. New equilibrium points

```python
# Points, demand & supply functions
p = np.arange(0.01, 5.0, 0.01)

plt.plot([280, 245, 190, 141, 135, 110, 95, 65, 58, 44, 21, 10], 
         [1, 1.25, 1.57, 1.81, 2.09, 2.45, 2.8, 3.19, 3.58, 3.85, 4.5, 5], 'ro')
plt.plot(Q_s(p), p, lw=2, c='blue')

plt.plot([5, 20, 51, 89, 120, 153, 180, 201, 215, 228, 240, 248],
         [1, 1.25, 1.57, 1.81, 2.09, 2.45, 2.8, 3.19, 3.58, 3.85, 4.5, 5], 'bo')
plt.plot(Q_d(p), p, lw=2, c='red')


# Supply with subsidy
for i in range(1, 5):
    p = np.arange(0.01 + 0.5 * i, 5.0, 0.01) # needed to avoid neg arg
    plt.plot(Q_d(p - 0.5 * i), p, lw=2)
#    idx = np.argwhere(np.diff(np.sign(Q_s(arg) - Q_d(p)) != 0).reshape(-1) + 0)
#    plt.plot(Q_s(p + 2 ** i)[idx], p[idx], 'ro')


# Legend
supply_patch = mpatches.Patch(color='b', label='Supply')
demand_patch = mpatches.Patch(color='r', label='Demand')
supply1_patch = mpatches.Patch(color='C0', label='Supply with subsidy in 0.5$')
plt.annotate('Equilibrium (s=0.5$)', xy=(150, 2.55), xytext=(147, 1),
            arrowprops=dict(facecolor='C0', shrink=0.05))
supply2_patch = mpatches.Patch(color='C1', label='Supply with subsidy in 1$')
plt.annotate('Equilibrium (s=1$)', xy=(168, 2.75), xytext=(115, 4.6),
            arrowprops=dict(facecolor='C1', shrink=0.05))
supply4_patch = mpatches.Patch(color='C2', label='Supply with subsidy in 1.5$')
plt.annotate('Equilibrium (s=1.5$)', xy=(185, 3.1), xytext=(160, 4.3),
            arrowprops=dict(facecolor='C2', shrink=0.05))
supply8_patch = mpatches.Patch(color='C3', label='Supply with subsidy in 2$')
plt.annotate('Equilibrium (s=2$)', xy=(203, 3.5), xytext=(190, 4),
            arrowprops=dict(facecolor='C3', shrink=0.05))
plt.legend(handles=[supply_patch, demand_patch, supply1_patch, supply2_patch, supply4_patch, supply8_patch])


# Plot scaling
plt.title('MMEEP lab1')
plt.xlabel('Quantity, N')
plt.ylabel('Price, $')
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 20
fig_size[1] = 10
plt.rcParams["figure.figsize"] = fig_size
plt.grid(True)
plt.xlim([0, 400])
plt.ylim([0, 6])
plt.show()
```
